## Processing one image

In practice, it's much more common to process many images at once. We'll start by processing one image at a time so we can break down what's happening.

From keras's image preprocessing module, let's use `load_img` and `img_to_array`:

```python
food = load_img('../data/yelp_mex_food_pics/adelitas_bad02.jpg')
food
```

```python
food_array = img_to_array(food)
```

Check the shape of the array to confirm that the image has three channels:

```python
food_array.shape
```

It's useful to know `plt.imshow()`. Note that `plt.imshow()` will require normalized pixel values:

```python
plt.imshow(food_array/255)
```
